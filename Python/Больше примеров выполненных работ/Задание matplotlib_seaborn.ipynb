1. Постройте круговую диаграмму для признака Rooms, иллюстрирующую количество квартир в процентах в зависимости от количества комнат. Сделайте сектор с наибольшим числом квартир выдвинутым.

```python
import matplotlib.pyplot as plt
```

```python
df.groupby(['Rooms'])['Id'].count()
```

```python
fig = plt.figure(figsize =(10, 7))

data = df.groupby(['Rooms'])['Id'].count()
explode = [0 if i != df.groupby(['Rooms'])['Id'].count().max() else 0.15 for i in df.groupby(['Rooms'])['Id'].count()]
plt.pie(data,
        explode=explode,
        autopct='%.1f%%',
        labels=data.index)

plt.title('Кол-во комнат в квартирах')
 
plt.show()
```

2. Постройте гистограмму целевой переменной Price. Оцените визуально, по какой цене продаётся наибольшее количество квартир.

```python
fig, ax = plt.subplots(figsize =(10, 7))

data = df['Price']
ax.hist(data, 
       bins = 50,
       edgecolor='white')

plt.xlabel("Цена квартиры")
plt.ylabel("Кол-во квартир по такой цене")
plt.title('Кол-во квартир по определенной цене')
 
plt.show()
```

```python
# Большинство квартир продаются по цене около 200000 у.д.е
```

3. Постройте диаграммы рассеяния для признаков Rooms, Square, HouseFloor, HouseYear в зависимости от целевой переменной Price в одной области fiagure(). Оцените визуально, есть ли среди них такие, на которых разброс точек близок к линейной функции

```python
import numpy as np
```

```python
fig, ax = plt.subplots(4, figsize=(20, 30))

ax[0].scatter(x = df['Price'], y = df['Rooms'], c = np.random.rand(10000))
ax[0].set_xlabel("Price")
ax[0].set_ylabel("Rooms")
ax[0].set_title('Диаграмма рассеивания - кол-во комнат от цены')

ax[1].scatter(x = df['Price'], y = df['Square'], c = np.random.rand(10000))
ax[1].set_xlabel("Price")
ax[1].set_ylabel("Square")
ax[1].set_title('Диаграмма рассеивания - площадь от цены')

ax[2].scatter(x = df['Price'], y = df['HouseFloor'], c = np.random.rand(10000))
ax[2].set_xlabel("Price")
ax[2].set_ylabel("HouseFloor")
ax[2].set_title('Диаграмма рассеивания - номер этажа от цены')

ax[3].scatter(x = df['Price'], y = df['HouseYear'], c = np.random.rand(10000))
ax[3].set_xlabel("Price")
ax[3].set_ylabel("HouseYear")
ax[3].set_title('Диаграмма рассеивания - год построения дома от цены')

plt.show()
```

```python
# Больше всего линейная зависимость виднеется на графиках с HouseYear (но его стоит посмотреть отдельно, 
# тк один выброс мешает визуальному анализу) и Square
```

4. Постройте ядерную оценку плотности целевой переменной Price. Оцените визуально, напоминает ли полученный график нормальное распределение. Постройте двумерную ядерную оценку плотности для целевой переменной Price и признака HouseFloor, затем оцените визуально на каких этажах и по какой цене продаётся основная масса квартир.

```python
import seaborn as sns
```

```python
plt.figure(figsize=(15, 5))

plt.title("Ядерная оценка плотности")

sns.kdeplot(data=df['Price'], lw=4, label="Ядерная оценка")

plt.legend()
plt.grid()
plt.show()
```

```python
# Да, нормальное распределение напоминает
```

```python
plt.figure(figsize=(15, 5))

sns.kdeplot(x = df['HouseFloor'], 
            y = df['Price'],
            cmap='Reds')
plt.title('2D ядерная оценка - параметры HouseFloar и Цена')
plt.grid()
plt.show()
```

```python
# Основная масса квартир продается по цене от 100000 до 200000 на 16-18 этажах
```

7. Постройте тепловую карту матрицы корреляции (df.corr()) признаков Rooms, Square, HouseFloor, HouseYear, Price. По ней определите, какие признаки являются зависимыми (у таких признаков коэффициент корреляции близок к единице).

```python
sns.heatmap(df[['Rooms', 'Square', 'HouseFloor', 'HouseYear', 'Price']].corr(), vmin = -1, vmax = +1, annot = True, cmap = 'coolwarm')
```

4. Выведите несколько изображений на экран используя инструментарий библиотеки matplotlib;

```python
import matplotlib.pyplot as plt

def print_faces(images, target, n):
    fig = plt.figure(figsize=(12,12))
    fig.subplots_adjust(left = 0,
                        right = 1,
                        bottom = 0,
                        top = 1,
                        hspace = 0.05,
                        wspace = 0.05
    )
    for i in range(n):
        p = fig.add_subplot(10, 10, i + 1, xticks = [], yticks = [])
        p.imshow(images[i], cmap = plt.cm.bone)
        p.text(0, 14, str(target[i]))
        p.text(0, 60, str(i))

print_faces(faces.images, faces.target, 100)
```

5. Разделите эти данные на тестовую и обучающую выборки;

```python
from sklearn.model_selection import train_test_split
```

```python
X = faces.data
y = faces.target

print(X.shape)
print(y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.2, 
                                                    random_state=3, 
                                                    #Одинаковая пропорция значений 'target' в тестовой выборке, что 
                                                    #и в самом faces.target
                                                    stratify=y)
```

6. Постройте модель классификатора метода опорных векторов для идентификации человека по изображению;

```python
from sklearn.svm import SVC
```

```python
classifier = SVC(kernel="linear") 
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

7. Оцените качество модели на тестовой выборке с помощью следующих метрик:
- достоверность предсказания (accuracy);
- точность (precision);
- полнота (recall);

```python
from sklearn import metrics
```

```python
metrics.accuracy_score(y_test, y_pred)
```

```python
metrics.precision_score(y_test, y_pred, average='weighted')
```

```python
metrics.recall_score(y_test, y_pred, average='weighted')
```

- "macro" просто вычисляет среднее значение двоичных показателей, придавая каждому классу одинаковый вес. В задачах, где редкие занятия тем не менее важны, макро-усреднение может быть средством выделения их производительности. С другой стороны, предположение, что все классы одинаково важны, часто неверно, так что макро-усреднение будет чрезмерно подчеркивать обычно низкую производительность для нечастого класса.
- "weighted" учитывает дисбаланс классов, вычисляя среднее значение двоичных показателей, в которых оценка каждого класса взвешивается по его присутствию в истинной выборке данных.
- "micro" дает каждой паре выборка-класс равный вклад в общую метрику (за исключением результата взвешивания выборки). Вместо того, чтобы суммировать метрику для каждого класса, это суммирует дивиденды и делители, составляющие метрики для каждого класса, для расчета общего частного. Микро-усреднение может быть предпочтительным в настройках с несколькими ярлыками, включая многоклассовую классификацию, когда класс большинства следует игнорировать.

8. Постройте кривую обучения - график зависимости тестовой и обучающей эффективности от размера обучающей выборки.

```python
for_x = []
for_y_accuracy = []
for_y_precision = []
for_y_recall = []

now_size = 0.2
delta = 0.01

while now_size < 0.8:
    for_x.append(now_size)
#     print(now_size)
    X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size= now_size, 
                                                    random_state=3,
                                                    stratify=y)
    classifier.fit(X_train, y_train)
    y_pred = classifier.predict(X_test)
    
    for_y_accuracy.append(metrics.accuracy_score(y_test, y_pred))
    for_y_precision.append(metrics.precision_score(y_test, y_pred, average='weighted'))
    for_y_recall.append(metrics.recall_score(y_test, y_pred, average='weighted'))
    
    now_size += delta
```

```python
fig, ax = plt.subplots(figsize=(20, 6))
ax.plot(for_x, for_y_accuracy)

ax.set_xlabel('Объем обучающей выборки (в долях)', 
           {'fontname':'Times New Roman'}, 
           fontweight='bold', 
           fontsize=16
          )
ax.set_ylabel('Accuracy', 
           {'fontname':'Times New Roman'}, 
           fontweight='bold', 
           fontsize=16
          )
ax.set_title('Изменение accuracy при уменьшении объема обучающей выборки', 
           {'fontname':'Times New Roman'}, 
           fontweight='bold', 
           fontsize=20
         )

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20, 6))
ax.plot(for_x, for_y_precision)

ax.set_xlabel('Объем обучающей выборки (в долях)', 
           {'fontname':'Times New Roman'}, 
           fontweight='bold', 
           fontsize=16
          )
ax.set_ylabel('Precision', 
           {'fontname':'Times New Roman'}, 
           fontweight='bold', 
           fontsize=16
          )
ax.set_title('Изменение precision при уменьшении объема обучающей выборки', 
           {'fontname':'Times New Roman'}, 
           fontweight='bold', 
           fontsize=20
         )

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20, 6))
ax.plot(for_x, for_y_recall)

ax.set_xlabel('Объем обучающей выборки (в долях)', 
           {'fontname':'Times New Roman'}, 
           fontweight='bold', 
           fontsize=16
          )
ax.set_ylabel('Recall', 
           {'fontname':'Times New Roman'}, 
           fontweight='bold', 
           fontsize=16
          )
ax.set_title('Изменение recall при уменьшении объема обучающей выборки', 
           {'fontname':'Times New Roman'}, 
           fontweight='bold', 
           fontsize=20
         )

plt.show()
```
